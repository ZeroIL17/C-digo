```python
# --- Paso 1 y 2: Eliminar columnas innecesarias de df_consumos ---
df_consumos = df_consumos.drop(columns=["Nombre Corto", "Archivo"])

# --- Paso 3: Agregar columna REGION a df_consumo_2024 ---

# 3.1 Comunas presentes en df_consumos
comunas_consumos = df_consumos["Comuna"].unique()

# 3.2 Localidades presentes en df_consumos
localidades_consumos = df_consumos["Localidad"].unique()

# 3.3 Eliminar filas en df_consumo_2024 cuya comuna no esté en df_consumos
df_consumo_2024 = df_consumo_2024[df_consumo_2024["DESCRIPCION_COMUNA"].isin(comunas_consumos)]

# 3.4 Eliminar filas en df_consumo_2024 cuya localidad no esté en df_consumos
df_consumo_2024 = df_consumo_2024[df_consumo_2024["NOMBRE_LOCALIDAD"].isin(localidades_consumos)]

# 3.5 y 3.6: Eliminar columnas CODIGO_LOCALIDAD y CODIGO_COMUNA
df_consumo_2024 = df_consumo_2024.drop(columns=["CODIGO_LOCALIDAD", "CODIGO_COMUNA"])

# 3.7 Asignar Región cruzando por comuna
df_comuna_region = df_consumos[["Comuna", "Región"]].drop_duplicates()
df_consumo_2024 = df_consumo_2024.merge(df_comuna_region, how="left",
                                        left_on="DESCRIPCION_COMUNA", right_on="Comuna").drop(columns=["Comuna"])

# --- Paso 4: Eliminar columna TIPO_SERVICIO ---
df_consumos = df_consumos.drop(columns=["Tipo Servicio"])
df_consumo_2024 = df_consumo_2024.drop(columns=["TIPO_SERVICIO", "DESCRIPCION_TIPO_SERVICIO"])

# --- Paso 5: Eliminar columnas TIPO_LIMITE y DESCRIPCION_TIPO_LIMITE ---
df_consumo_2024 = df_consumo_2024.drop(columns=["TIPO_LIMITE", "DESCRIPCION_TIPO_LIMITE"])

# --- Paso 6: Eliminar columnas RANGO y RANGO_ID ---
df_consumo_2024 = df_consumo_2024.drop(columns=["RANGO", "RANGO_ID"])

# --- Paso 7: Eliminar columna M3_AS ---
df_consumos = df_consumos.drop(columns=["M3 As"])
df_consumo_2024 = df_consumo_2024.drop(columns=["M3_AS"])

# --- Paso 8: Agregar Clientes Totales ---

# 8.1 En df_consumo_2024
df_consumo_2024["CLIENTES_TOTALES"] = df_consumo_2024["CLIENTES_DIRECTOS"] + df_consumo_2024["CLIENTES_INDIRECTOS"]

# 8.2 En df_consumos: Cruzar con df_clientes
cols_join = ["Comuna", "Tipo Cliente", "Razon Social", "Año", "Mes"]
df_consumos = df_consumos.merge(
    df_clientes[cols_join + ["Clientes"]],
    on=cols_join,
    how="left"
)
```

```python
# === df_consumos ===
# Crear columna "Fecha" en df_consumos a partir de Año y Mes
df_consumos["Fecha"] = pd.to_datetime(
    df_consumos[["Año", "Mes"]].rename(columns={"Año": "year", "Mes": "month"}).assign(day=1)
)

# === df_consumo_2024 ===

# 1. Extraer la fecha desde PERIODO_INFORMADO
df_consumo_2024["Fecha"] = pd.to_datetime(df_consumo_2024["PERIODO_INFORMADO"], format="%Y%m")
df_consumo_2024.drop(columns=["PERIODO_INFORMADO"], inplace=True)
df_consumo_2024["Año"] = df_consumo_2024["Fecha"].dt.year
df_consumo_2024["Mes"] = df_consumo_2024["Fecha"].dt.month

# 2. Eliminar columna TIPO_CLIENTE y renombrar DESCRIPCION_TIPO_CLIENTE a "Tipo Cliente"
df_consumo_2024.drop(columns=["TIPO_CLIENTE"], inplace=True)
# Si ya existe "Tipo Cliente", este paso podría omitirse, pero se asegura aquí:
if "DESCRIPCION_TIPO_CLIENTE" in df_consumo_2024.columns:
    df_consumo_2024.rename(columns={"DESCRIPCION_TIPO_CLIENTE": "Tipo Cliente"}, inplace=True)

# 3. Eliminar CLIENTES_DIRECTOS e INDIRECTOS y renombrar CLIENTES_TOTALES
df_consumo_2024.drop(columns=["CLIENTES_DIRECTOS", "CLIENTES_INDIRECTOS"], errors="ignore", inplace=True)
df_consumo_2024.rename(columns={"CLIENTES_TOTALES": "Clientes"}, inplace=True)

# 4-8. Renombrar columnas restantes
df_consumo_2024.rename(columns={
    "RUT": "Rut",
    "RAZON_SOCIAL": "Razon Social",
    "NOMBRE_LOCALIDAD": "Localidad",
    "DESCRIPCION_COMUNA": "Comuna",
    "M3_AP": "M3 Ap"
}, inplace=True)

# === Unir ambos dataframes ===
# Asegurar que ambas tablas tengan las mismas columnas y en el mismo orden
# (puedes ajustar manualmente si alguna otra diferencia aparece)
df_consumo_2024 = df_consumo_2024[df_consumos.columns]  # asegura orden y columnas coincidentes

# Concatenar para formar df_2012_2024
df_2012_2024 = pd.concat([df_consumos, df_consumo_2024], ignore_index=True)
```

```python
# Renombrar columna de consumo en ambos DataFrames a un nombre común
df_agrupado = df_agrupado.rename(columns={'Consumo Total M3 Ap': 'M3_Ap'})
df_agrupado_2024 = df_agrupado_2024.rename(columns={'Consumo Total M3_AP': 'M3_Ap'})

# Unir ambos DataFrames en uno solo
df_mensual = pd.concat([df_agrupado, df_agrupado_2024], ignore_index=True)

# Ordenar por fecha por si no vienen en orden
df_mensual = df_mensual.sort_values('Fecha').reset_index(drop=True)
```

```python
# 1. Unir df_mensual y df_refac_agrupado por la columna Fecha
df_mensual = df_mensual.merge(df_refac_agrupado, on='Fecha', how='left')

# 2. Rellenar posibles NaN de M3_Refacturado con 0
df_mensual['M3_Refacturado'] = df_mensual['M3_Refacturado'].fillna(0)

# 3. Sumar M3_Ap + M3_Refacturado en una nueva columna (o sobrescribir la existente)
df_mensual['M3_Ap'] = df_mensual['M3_Ap'] + df_mensual['M3_Refacturado']

# 4. (Opcional) Eliminar la columna M3_Refacturado si ya no se necesita
df_mensual = df_mensual.drop(columns=['M3_Refacturado'])
```
